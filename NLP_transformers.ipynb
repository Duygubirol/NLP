Pipeline

This pipeline is doing 3 things for us:

Preprosessing the text such as tokenizer

It feed the preprocessing text to the model and apply model 

Postprocessing (it will show us the result. The expectation)

```python
from transformers import pipeline
```

```python
# pretrained models
classifier = pipeline("sentiment-analysis")
res1 =classifier("I am too tired to learn too many things at the same time but I know I can do it")
print(res1)
```

```python
classifier = pipeline("sentiment-analysis") # default model used
```

Text Generation


```python
generator = pipeline ("text-generation", model ="distilgpt2")

res2 = generator(
     "In this example, we will learn how to",
     max_length=30,
     num_return_sequences=2,
     )
print(res2)
```

Identification of candidate label

```python
classifier= pipeline("zero-shot-classification")
```

```python
res3= classifier(
    "This is the hardest course ",
    candidate_labels=["education","politics","business"]
)
print(res3)
```
